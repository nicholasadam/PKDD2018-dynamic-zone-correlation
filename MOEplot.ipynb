```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats
```

```python
tod=np.loadtxt('tod.txt')
zonalArrival=np.loadtxt('arrival.txt')
```

```python
# data to plot
n_groups = tod.shape[1]
predict = tod[0]/tod[0].sum()
reference = tod[1]/tod[1].sum()
 
# create plot
fig, ax = plt.subplots()
index = np.arange(1,n_groups+1)
bar_width = 0.3
opacity = 0.8
 
rects1 = plt.bar(index, predict, bar_width,
                 alpha=opacity,
                 color='y',
                 label='HPSS')
 
rects2 = plt.bar(index + bar_width, reference, bar_width,
                 alpha=opacity,
                 color='b',
                 label='Ground Truth')
 
plt.xlabel('Time of day (Hour)')
plt.ylabel('Trip arrival frequency')
plt.title('(a) Temporal Distribution of Trip Arrival Estimation')
plt.xticks(index[::3])
plt.legend()
plt.tight_layout()
plt.show()
ax.set_aspect(aspect=90)
#fig.set_size_inches(4.8,3, forward=True)
fig.savefig('TOD.png', dpi=300)
```

```python
reference=np.load('OD.npy').sum(1)
predict=np.load('predict.npy').sum(1)
#predict=predict*0.6+reference*0.4
```

```python
fig, ax = plt.subplots()
fit = np.polyfit(reference, predict, deg=1)
line1,=ax.plot(reference, fit[0] * reference + fit[1], color='red',\
               label='Regression line: Y= %.2f+%.2f*x' % (fit[1],fit[0]))
line2,=ax.plot(reference, reference, color='green',label='Line Y=x')
ax.scatter(reference, predict,s=10)
ax.set_xlim([0, 8000])
ax.set_ylim([0, 8000])
plt.xlabel('Reference zonal trip arrivals')
plt.ylabel('Predicted zonal trip arrivals')
plt.title('(b) Predicted versus Reference Zonal Trip Arrivals')
ax.set_aspect(aspect=0.5, adjustable='box')
plt.legend(handles=[line1, line2])
fig.show()
fig.savefig('regression.png', dpi=300)
```
